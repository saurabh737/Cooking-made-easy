Calculating unique ingredients

```python
def calc_unique(column):
    unique_item = set()
    for row in column:
        row = ast.literal_eval(row)
        for i in row:
            unique_item.add(i)
            
    return unique_item
```

```python
unique_ingredients = calc_unique(raw_recipes['ingredients'])
len(unique_ingredients)
```
